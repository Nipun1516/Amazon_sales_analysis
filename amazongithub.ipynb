```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('amazon.csv')
df
```

# now we will take the inforamtion of the data sheet

```python
df.info()
```

## Here we can see that in 19 & 20 index there are emplty values so we sill drop it 
## here we use inplace=True to save changes & for delete particular row 

```python
df.drop(['New','PendingS'],axis=1,inplace=True)
```

```python
df.info()
```

# here we can seee that we have delete the null row 
## Now we will check the null values 

```python
pd.isnull(df).sum()
```

## Now we will clean the null value 


```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

# we can see that we have clean all the null values form data 
## if we want know the colums name 
## and want to know the total rows and columns of data sheet

```python
df.columns
#df.shape
```

## Now we will change the data type of 'ship-postal-code' into integer

```python
df['ship-postal-code']=df['ship-postal-code'].astype('int')
df['ship-postal-code'].dtype
```

## we have finaly change the datatype into int

# Now we will create a chart for 'SIZE'


```python
ax=sns.countplot(x='Size',data=df)
```

## Now if we want to see te count so we will run the for loop 

```python
ax=sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

# Groupby now we will use the group by function 
## the groupby() function in pandas is used to database one or more column in a dataframe

```python
group=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=group)
```

# courier status 

```python
sns.countplot(x='Courier Status',data=df,hue='Status')
```

# histogram

```python
df['Size'].hist()
```

```python
df['Category']=df['Category'].astype(str)
col=df['Category']
plt.figure(figsize=(10,5))
plt.hist(col,bins=30,edgecolor='Black')
plt.xticks(rotation=90)
plt.show()


# bin are sie or bar, edgecolor are color of edge, rotaiton90 is name of X axis which are rotate
```

# pie chart 

```python
b2b=df['B2B'].value_counts()
plt.pie(b2b,labels=b2b.index)
plt.show()
```

# scatter plot

```python
x_data=df['Category']
y_data=df['Size']
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('scatter plot')
plt.show()
```
